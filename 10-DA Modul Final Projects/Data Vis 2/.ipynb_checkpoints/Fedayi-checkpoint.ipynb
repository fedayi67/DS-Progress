## CODING TASK #1. PLOT PIE CHART USING MATPLOTLIB

- **The plot method on Pandas Series and DataFrames is just a simple wrapper around plt.plot():**
- **Define a Pandas Dataframe with all crypto allocation in a given portfolio**
- **Note that total sum = 100%** 
- **Use matplotlib to plot a pie chart** 

```python
import numpy as np
import pandas as pd

import matplotlib as mpl
import matplotlib.pyplot as plt

%matplotlib inline
import seaborn as sns
```

```python
# Define a Pandas Dataframe with all crypto allocation in a given portfolio
# Note that total sum = 100%

crypto_DF = pd.DataFrame(data = {"allocations":[60,15,7,10,8]}, index = ["BTC","ETH",'LTC','ADA','XRP'])
crypto_DF
```

```python
# we can 
plt.figure(figsize = (7,7))
plt.pie(crypto_DF.allocations, labels = crypto_DF.index, autopct = "%.1f%%",frame = True,labeldistance = 0.75)
plt.legend()
plt.show();
```

```python
# Use matplotlib to plot a pie chart
# we can plot dirrectly by DataFrame or Series.
crypto_DF.plot.pie(figsize = (8,8), shadow = True, autopct = "%1.1f%%",colors =["#FFD700","green","grey","orange","blue"], subplots = True)
plt.title("Crypto Portfolio Pie Chart")
plt.axis('equal'); # this line of code ensures a circular pie chart.
```

**PRACTICE OPPORTUNITY #1:**
- **Assume that you became bullish on XRP and decided to allocate 60% of your assets in it. You also decided to equally divide the rest of your assets in other coins (BTC, LTC, ADA, and ETH). Change the allocations and plot the pie chart.**
- **Use 'explode' to increase the separation between XRP and the rest of the portfolio (External Research is Required)**

```python
# total allocations 
total_allocation = crypto_DF.allocations.sum()
total_allocation
```

```python
# assigning 60% of your asset or allocation to XRP
crypto_DF.allocations.XRP = total_allocation*0.60 # crypto_DF["allocations"]["XRP"]
```

```python
# divide the remaining of your assets equally upon the rest
remaining_per_coin = (total_allocation - crypto_DF.allocations.XRP) / (len(crypto_DF.allocations) - 1)
remaining_per_coin
```

```python
# update the percentages
for i in ['ETH', 'LTC', 'BTC', 'ADA']:
    crypto_DF.allocations[i] = remaining_per_coin # i has to be written inside the brackets, otherwise it'll look for i inside 
    # the datafram
```

```python
crypto_DF
```

```python
# use MATPLOTLIB to plot a pie chart with explode separation on XRP
plt.figure(figsize = (8,8))
plt.pie(crypto_DF.allocations, labels = crypto_DF.index, autopct = "%.1f%%",frame = True,labeldistance = 0.75,explode = [0,0,0,0,0.1])
plt.legend()
plt.show();
```

## CODING TASK #5. PLOT SCATTERPLOT & COUNTPLOT USING SEABORN
- **Seaborn is a visualization library that sits on top of matplotlib**

- **Check out the head of the dataframe**

- **Let's print out countplot to know how many samples belong to class #0 and #1**

```python
# import libraries 
import pandas as pd # Import Pandas for data manipulation using dataframes
import numpy as np # Import Numpy for data statistical analysis 
import matplotlib.pyplot as plt # Import matplotlib for data visualisation
import seaborn as sns # Statistical data visualization
```

```python
# Import Cancer data 
cancer_df = pd.read_csv('cancer.csv')
```

```python
cancer_df.tail()
```

```python
# Plot scatter plot between mean area and mean smoothness
plt.rcParams['figure.facecolor'] = "#ffffe6"
plt.rcParams['axes.facecolor'] = "#ffffe6"
sns.scatterplot(data=cancer_df, hue="target",x="mean area", y="mean smoothness", color="r")
plt.title('Mean Area and Mean Smoothness')
plt.show()
```

Cancer cases tend to concentrate in an area characterized by a mean area of less than 700 and a mean smoothness of less than 0.12.

For every cell nucleus, ten real-valued features are calculated:

a) Radius: Mean of distances from the center to points on the perimeter.

b) Texture: Standard deviation of gray-scale values.

c) Perimeter.

d) Area.

e) Smoothness: Local variation in radius lengths.

f) Compactness: (Perimeter^2 / Area) - 1.0.

g) Concavity: Severity of concave portions of the contour.

h) Concave points: Number of concave portions of the contour.

i) Symmetry.

j) Fractal dimension: "Coastline approximation" - 1.

These initial 10 features (a...j) are expanded to 30 through feature engineering. The mean, standard error, and "worst" or largest (mean of the three largest values) of these features are computed for each image, resulting in a total of 30 features. For example, field 3 represents Mean Radius, field 13 represents Radius SE, and field 23 represents Worst Radius.

```python
plt.rcParams['figure.facecolor'] = "#ffffe6"
plt.rcParams['axes.facecolor'] = "#ffffe6"
plt.figure(figsize=(8,5))
plt.title('Number of Class 0 and 1')
sns.countplot(x=cancer_df["target"],palette='inferno')
plt.show()
```

## CODING TASK #6. PLOT PAIRPLOT, DISPLOT, AND HEATMAPS/CORRELATIONS USING SEABORN

- **Plot the pairplot**
- **Strong correlation between the mean radius and mean perimeter, mean area and mean primeter**
- **plot the distplot** 
- **Displot combines matplotlib histogram function with kdeplot() (Kernel density estimate)**
- **KDE is used to plot the Probability Density of a continuous variable.** 

```python
# heatmap graph with seaborn
corr = cancer_df.corr()
mask = np.zeros_like(corr) # corr kalibini 0 ile dolduruyor
mask[np.triu_indices_from(mask)] = True # mask'n st geni iin indisleri dndrr.ust ucgene 0 doldurdu
with sns.axes_style("white"):
    f, ax = plt.subplots(figsize=(8, 8))
    ax = sns.heatmap(corr,mask=mask,square=True,linewidths=.8,cmap="autumn_r") 
```

**PRACTICE OPPORTUNITY #6:**
- **Plot two separate distplot for each target class #0 and target class #1**


```python
plt.figure(figsize=(8, 4))
sns.displot(cancer_df[cancer_df['target'] == 0]['mean radius'], kde=True, color='blue')
plt.title(' Target Class 0')
plt.show()

plt.figure(figsize=(8, 4))
sns.displot(cancer_df[cancer_df['target'] == 1]['mean radius'], kde=True, color='purple')
plt.title(' Target Class1')
plt.show()
```
