# 1. Covariance and Correlation

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

# 2. SciPy

```python
Ekran_sure = [3,5,2,0.5,5,3,1,4,3,4]
AGNO = [2.7,2.2,3.3,3.4,2,3,3.6,2.4,3.3,2.6]
```

https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.linregress.html#scipy.stats.linregress

```python
# Pearson r katsayisi hesabi ... stats modulu ile

stats.pearsonr(Ekran_sure, AGNO)
# statistic = r
# in interviews pvalue can be asked.
```

```python
# x ve y'nin yer degisimi

stats.pearsonr(AGNO,Ekran_sure)
```

```python
# spearman correlation katsayisi hesabi

stats.spearmanr(Ekran_sure, AGNO) 
# non-parametric test: 
# normal dagtilima uygun olmayan veriler uzerine calismak icin
```

# 3. Numpy

```python
# Covariance hesabi

np.cov(AGNO, Ekran_sure)
# -0.78 is covariance, a two by two matrix, think of it as correlation matrix.
```

```python
# correlation hesabi .... kosegen degerleri r degeridir.

np.corrcoef(AGNO, Ekran_sure)
```

# 4. Linear Regression 

```python
# 

stats.linregress(Ekran_sure, AGNO)
```

```python
# yukaridaki degerleri alttaki variable'lara atayalim.

slope, intercept, r_value, p_value, std_err = stats.linregress(Ekran_sure, AGNO)
```

```python
slope
```

```python
intercept
```

```python
r_value
```

```python
# R2 hesabi

R_squared = r_value**2
R_squared
```

```python
# let's predict

x_p = np.array([0,1,2,3,4,5])
y_p = intercept + slope*x_p
```

```python
x_p
```

```python
y_p
```

```python
# linspace ile deger uretelim .. arasinda

x_new = np.linspace(0,5,100)
y_new = intercept + slope * x_new
```

```python
x_new
```

```python
plt.figure(figsize=(8,6))
ax = plt.axes()     #
ax.scatter(Ekran_sure, AGNO)
ax.plot(x_new, y_new)
ax.set_xlabel("Ekran Sure")
ax.set_ylabel("AGNO")
ax.axis("tight")
plt.show()
```
