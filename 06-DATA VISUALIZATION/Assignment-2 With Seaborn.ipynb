## 2: Plot 12 most dangerous cities

```python
kill.head()
```

```python
kill.city.value_counts().head(12)
```

```python
cty = kill.value_counts("city").nlargest(12) # the difference between the results of this code and the code above is :
# that the code above brings out the output that comes first alphabetically, in order to confirm let's try with first 15 O/Ps
cty
```

```python
kill.city.value_counts().head(15)
```

```python
kill.value_counts("city").nlargest(15) # we can clearly see here that this line of code follows a certain type of pattern
# whereas the code above gives the output in alphabetical order
# it looks like the second code keeps counting, and if two cities have same number of counts, it shows the one that finishes first as first entity.
```

```python
pd.value_counts(kill.city)[:12] # solution 3
```

```python
yvalue = cty.values
yvalue
```

```python
xindex = cty.index
xindex
```

```python
cty.values.tolist()
```

```python
fig, ax = plt.subplots(figsize = (10,6))

sns.barplot(x = xindex, y = yvalue)
```

```python
fig, ax = plt.subplots(figsize = (10,6))

sns.barplot(x = xindex, y = yvalue)
ax.set_xticklabels(cty.index, color = "b", fontsize = 8);
```

```python
fig, ax = plt.subplots(figsize = (10,6))

sns.barplot(x = xindex, y = yvalue)
ax.set_xticklabels(cty.index, color = "b", fontsize = 8, rotation = 45);
```

```python
fig, ax = plt.subplots(figsize = (10,6))

sns.barplot(x = xindex, y = yvalue)
ax.set_xticklabels(cty.index, color = "b", fontsize = 8, rotation = 45)
ax.set_title("Most Dangerous Cities", color = "red");
```

```python
# second way
cty.values.tolist()
plt.figure(figsize=(8.4,5))
sns.barplot(x=cty.index,
            y=cty.values)
plt.xticks(rotation=45,
           color="blue",
           fontsize= 8)
plt.yticks(fontsize=8)
plt.title(" Most Dangerous Cities",
          color="red")
plt.show()
```

```python
# with seaborn
fig, ax = plt.subplots(figsize=(10,6))
sns.barplot(x=xindex, y=yvalue)
ax.set_xticklabels(cty.index, color="b", fontsize=8, rotation= 45)
ax.set_title("Most Dangerous City", color="r");
```

```python
cty
```

```python
cty.index
```

```python
# with seaborn countplot
plt.figure(figsize=(6,4))
sns.countplot(x="city",
              data=kill,
              order=cty.index)
plt.xticks(rotation=45,
           fontsize=7)
plt.yticks(fontsize=7)
plt.title("Most Dangerous Cities",
          color='blue')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,5))
sns.countplot(x="city", data=kill, order=xindex)
ax.set_xticklabels(cty.index, color="b", fontsize=8, rotation= 45)
ax.set_title("Most Dangerous City", color="r");
```

## 3: Plot 20 most dangerous states

```python
kill_state = kill.state.value_counts().head(20)
kill_state
```

```python
x_index = kill_state.index
x_index
```

```python
y_values = kill.state.value_counts().head(20).values
y_values
```

```python
# with seaborn barplot
fig, ax = plt.subplots(figsize=(8,5))
sns.barplot(x= x_index, y= y_values)
ax.set_xticklabels(kill_state.index, color="b", fontsize=8, rotation= 45)
ax.set_title("Most Dangerous States", color="r");
```

```python
# with seaborn barplot
```

```python
# with seaborn countplot

fig, ax = plt.subplots(figsize=(8,5))
sns.countplot(x= "state", data=kill, order=x_index)
ax.set_xticklabels(kill_state.index, color="b", fontsize=8, rotation= 45)
ax.set_title("Most Dangerous States", color="r");
```
