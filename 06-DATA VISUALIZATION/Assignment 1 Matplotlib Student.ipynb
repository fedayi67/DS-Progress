## Kütüphaneleri import edin

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Değişkenlerimizi tanımlayalım

m değerini output a göre numpy kütüphanesi yardımı ile oluşturalım

```python
m = np.linspace(0,10,11) # endpoint in linspace is True
m 

#np.arange(11,dtype = float)
```

```python
m
```

```python
c = 3* 10**8
```

```python
E = m* c**2
```

```python
E
```

## Oluşturduğumuz dokümanları plotlayalım

#### PLOTTING E=mc^2

```python
plt.figure(figsize = (4,3))

plt.plot(m, E, color = "blue", lw = 3)
plt.title("E = mc^2")
plt.xlabel("M in g")
plt.ylabel("E in J"); # here  the semicolun helps not creating any detail related texts
```

```python
plt.figure(figsize = (6,4))

plt.plot(m, E, color = "orange", lw = 2)
plt.title("E = mc^2")
plt.xlabel("M in g")
plt.ylabel("E in J");
```

## Alttaki görseli oluştururken farklı bir değişken tanımlamayın. Bence eksenlerden birinin logaritmik hesaplanması gerekebilir:)

## Ayrıca axes de oluşan gridler bir style değil bence matplotlib library de gridleri araştırmanız gerekir

```python
plt.figure(figsize = (4,3))

plt.plot(m, E, color = "green", lw = 2.5)
plt.title("E = mc^2")
plt.xlabel("M in g")
plt.ylabel("E in J")
plt.yscale("log")
plt.grid(which='both',axis='y');
```

```python
plt.figure(figsize = (5, 4))
plt.semilogy(m, E, color = "green", linewidth=2)     # semilogy() y eksenini günlük ölçeklendirmeye dönüştürmek için kullanılır
                                                     # linewidth=2 line kalinligini ayarlar
plt.xlabel("M in g", fontsize=10)
plt.xticks(fontsize=10)  # x eksen deki elemanlarin boyutunu ayarlar
plt.xlim([0, 10])       # xlim x axis e limit koyar
plt.ylabel("E in j", fontsize=10)
plt.yticks(fontsize=8)  #y eksen deki elemanlarin boyutunu ayarlar
plt.title("E = mc^2", fontsize=12)
plt.grid(which='both', axis="y")
plt.show()
```

## Aşağıdaki değişkenlere göre ilgili platlamaları yapın

```python
labels = ["1 Mo", "3 Mo", "6 Mo", "1 Yr", "2 Yr", "3 Yr", "5 Yr", "7 Yr", "10 Yr", "20 Yr", "30 Yr"]
july16_2007 = [4.75, 4.98, 5.08, 5.01, 4.89, 4.89, 4.95, 4.99, 5.05, 5.21, 5.14]
july16_2020 = [0.12, 0.11, 0.13, 0.14, 0.16, 0.17, 0.28, 0.46, 0.62, 1.09, 1.31]
```

```python
plt.figure(figsize = (6,4))

plt.plot(labels,july16_2007,label = "july16_2007")
plt.plot(labels,july16_2020,label = "july16_2020")
plt.legend()
```

```python
fig = plt.figure(figsize = (5,4))

ax = fig.add_axes([0,0,1,1])
ax.plot(labels,july16_2007,label = "july16_2007")
ax.plot(labels,july16_2020,label = "july16_2020")

ax.legend(loc = (1.01,0.5));
```

## Alttaki görselleri araştırarak yapmaya çalışalım. Zihnimizi ve parmaklarımızı zorlayalım:)

#### Bir hatırlatma "fig, ax = plt.subplots" syntax i size cevabı bulmanızda yardımcı olur. 

#### ayrıca shift + tab+ tab yapmayı unutmayın ve içindeki parametreleri denemeyi unutmayın.

```python
fig, ax = plt.subplots(2,1)
ax[0].plot(labels,july16_2007,label="july16_2007")
ax[0].set_title("july16_2007")
ax[1].plot(labels,july16_2020,label="july16_2020")
ax[1].set_title("july16_2020")
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(2, 1)
ax[0].plot(labels, july16_2007, 'b')
ax[0].set_title('july16_2007')
ax[1].plot(labels, july16_2020, 'b')
ax[1].set_title('july16_2020')
plt.tight_layout();
```

```python
fig, ax = plt.subplots(2, 1, figsize=(7.3, 5.5))
ax[0].plot(labels,july16_2007, label = "july16_2007", lw=2)
ax[0].set_title("july16_2007", fontsize=9)
ax[0].tick_params(axis='x', labelsize=7.5)
ax[0].tick_params(axis='y', labelsize=7.5)
ax[1].plot(labels,july16_2020, label = "july16_2020", lw=2)
ax[1].set_title("july16_2020", fontsize=9)
ax[1].tick_params(axis='x', labelsize=7.5)
ax[1].tick_params(axis='y', labelsize=7.5)
plt.tight_layout();
```

## Bu görselleri alabilenler için alkışı borç bilirim

```python
fig, ax = plt.subplots(figsize=(8.5, 5))
ax.plot(labels, july16_2007, color='blue', label='July 16, 2007')
ax.set_ylabel('July 16 2007', color='blue', fontsize = 14)
ax2 = ax.twinx()
ax2.plot(labels, july16_2020, color='red', label='July 16, 2020')
ax2.set_ylabel('July 16 2020', color='red', fontsize = 14)
plt.show()
```

```python
fig, ax1 = plt.subplots(figsize = (8, 4))


ax1.set_ylabel("july16 2007", color="blue")
ax1.plot(labels,july16_2007, color = "blue",lw=2)

ax2 = ax1.twinx()

ax2.set_ylabel("july16 2020", color="red") 
ax2.plot(labels,july16_2020, color = "red",lw=2);
```

```python
fig, ax1 = plt.subplots(figsize = (9,5))
ax1.plot(labels, july16_2007, color='b', linewidth=3)
ax1.set_ylabel("july 16 2007",size = 12, color = "b")
ax1.spines["left"].set_color("b")   # bu calismiyor yazikki.
ax1.spines['left'].set_linewidth(1.5) # sol eksenin kalinligini ayarladik
ax2 = ax1.twinx()
ax2.plot(labels,july16_2020, color='r', linewidth=3)
ax2.set_ylabel('july 16 2020', color = "r", size = 12)
ax2.spines['right'].set_color('red') # sagdaki y eksenini red yapiyor.
ax2.spines['right'].set_linewidth(1.5) # sag eksenin kalinlgini ayarladik
ax2.set_title('Yield Curve Comparison: July 16, 2007 vs. July 16, 2020', color="purple", size = 16)
plt.show()
```
